```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Heart Disease data.csv")
```

```python
data
```

```python
# Data Inspection
print("First few rows of the dataset:")
print(data.head())

print("\nSummary of the dataset:")
print(data.info())

print("\nDescriptive statistics of the dataset:")
print(data.describe())
```

```python
# Checking for missing values
print("\nMissing values in the dataset:")
print(data.isnull().sum())
```

```python
# Summary Statistics
print("\nSummary statistics for categorical variables:")
print(data['sex'].value_counts())
print(data['age'].value_counts())
print(data['restecg'].value_counts())
```

```python
# Distribution of heart disease rates
plt.figure(figsize=(8, 6))
sns.countplot(x='restecg', data=data)
plt.title('Distribution of Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.show()
```

```python
# Heart disease rates by gender
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=data)
plt.title('Heart Disease by Gender')
plt.xlabel('sex')
plt.ylabel('Count')
plt.show()
```

```python
#Heart disease rates by age groups
plt.figure(figsize=(12, 8))
sns.histplot(data=data, x='age')
plt.title('Heart Disease by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
